```python
# Part of this code has been solved with the help of ChatGPT. 

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Defining the matrix A, vector b, and constant c
A = np.array([[13, 5], [5, 7]])
b = np.array([1, 1])
c = 2

# Defining the objective function f(x)
def f(x):
    return x.T @ A @ x + b.T @ x + c

# Defining the constraints g(x) = 0 and h(x) = 0
def g(x):
    return 2 * x[0] - 5 * x[1] - 2

def h(x):
    return x[0] + x[1] - 1
```

```python
# Creating the constraint dictionaries. "eq" means the constraint should be equal to zero. "fun" indicates it is a function used.
constraint_g = {'type': 'eq', 'fun': g}
constraint_h = {'type': 'eq', 'fun': h}
constraints = [constraint_g, constraint_h]

# Initial guess
x0 = [0, 0]

# Finding the solution
solution = minimize(f, x0, constraints=constraints)
print(solution.x)
```

```python
# Plotting all the graphs in one plot
x1_vals = np.linspace(-1, 2, 100)
x2_vals = np.linspace(-1, 2, 100)
X1, X2 = np.meshgrid(x1_vals, x2_vals)
F_vals = np.array([[f(np.array([x1, x2])) for x1 in x1_vals] for x2 in x2_vals])
G_vals = np.array([[g(np.array([x1, x2])) for x1 in x1_vals] for x2 in x2_vals])
H_vals = np.array([[h(np.array([x1, x2])) for x1 in x1_vals] for x2 in x2_vals])

plt.contour(X1, X2, F_vals, levels=50, cmap='viridis')
plt.contour(X1, X2, G_vals, levels=[0], colors='red', linestyles='dashed')
plt.contour(X1, X2, H_vals, levels=[0], colors='blue', linestyles='dashed')

# Scatter plot for the optimal point
plt.scatter(solution.x[0], solution.x[1], color='green', label='Optimal Point', s=100)
plt.title('Contours of f(x) with constraints g(x) and h(x)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar(label='f(x)')
plt.legend()
plt.show()
```
