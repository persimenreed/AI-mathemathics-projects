## Imports

Importing the dataset, seaborn for heatmap and pandas to handle the dataframe

```python
from sklearn.datasets import load_iris
from seaborn import heatmap as sns_heatmap
import pandas as pd
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

#### Calculating and printing feature wise mean

```python
feature_wise_mean = df.mean()
print(feature_wise_mean)
```

#### Calculating and printing correlation matrix

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

#### Calculating and printing variance convariance matrix

```python
variance_covariance_matrix = df.cov()
```

```python
print(variance_covariance_matrix)
```

#### Heatmap for correlation matrix

```python
sns_heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

#### Heatmap for variance covariance matrix

```python
sns_heatmap(variance_covariance_matrix, annot=True, cmap='coolwarm')
```
