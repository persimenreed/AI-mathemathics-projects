```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x=[132,129,120,113.2,105,92,84,83.2,88.4,59,80,81.5,71,69.2]
y=[46,48,51,52.1,54,52,59,58.7,61.6,64,61.4,54.6,58.8,58]
```

```python
x = np.array(x).reshape(-1, 1)
y = np.array(y)
regression = linear_model.LinearRegression()
regression.fit(x, y)
```

```python
# Predictions (y values of the regression line)
y_pred = regression.predict(x)

slope = regression.coef_[0]
intercept = regression.intercept_

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Plot only the regression line
plt.plot(x, y_pred, color="blue", linewidth=3)
plt.scatter(x, y, color="black", linewidth=3)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear Regression Line")
plt.show()
```
