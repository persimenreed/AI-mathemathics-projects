## Imports and data preparation
Importing the needed libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
## Loading the dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

## Only selecting one feature to vizualize
diabetes_X = diabetes_X[:, np.newaxis, 2]

## Initializing the linear regression object
regression = linear_model.LinearRegression()

## Training the model on the given dataset
regression.fit(diabetes_X, diabetes_y)

## Predicting data after being trained
diabetes_y_pred = regression.predict(diabetes_X)
```

```python
## Printing MSE and coefficient of determination
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y, diabetes_y_pred))
print("R2 coefficient: %.2f" % r2_score(diabetes_y, diabetes_y_pred))

## Plotting the original data together with the predicted data
plt.scatter(diabetes_X, diabetes_y, color="black", label="Original data")
plt.plot(diabetes_X, diabetes_y_pred, color="blue", linewidth=3, label="Regression line")

plt.legend()
plt.show()
```
